```python
import json, os, re
from typing import List
from tqdm import tqdm

response_dir = "../data/responses/meditron2-70b.jsonl"
hash_to_name = {}
meditron = {}
gold_prompts_dict = {}
                                   
def normalize(s):
    return re.sub(r'[^a-zA-Z0-9]', '', s.replace('\n', '')).lower().strip().replace('theansweris', '')

def read_result_json(path: str):
    """
    Read the raw result file.

    Args:
        path (str): Path to the raw result file.

    Returns:
        List[dict]: Contents of the raw result file.
    """
    responses = []
    with open(path, "r") as raw_result:
        for line in raw_result:
            responses.append(json.loads(line))
    return responses

meditron2 = read_result_json(response_dir)
gold = read_result_json("../data/exomiser-gold.jsonl")

normalize(gold[0]["prompt"]) == normalize(meditron2[0]["prompt"])
count = 0
for med, g in zip(meditron2, gold):
    count += 1
    assert normalize(med["prompt"]) == normalize(g["prompt"]), f"Prompt mismatch {count}: {med['prompt']} vs {g['prompt']}"
    med["id"] = g["id"]
    med["gold"] = g["gold"]
    med["response"] = med.pop("output")
```

```python
with open("../data/responses/meditron-70b.jsonl", "w") as f:
    for line in tqdm(meditron2, total=len(meditron2)):
        f.write(json.dumps(line) + "\n")
```
